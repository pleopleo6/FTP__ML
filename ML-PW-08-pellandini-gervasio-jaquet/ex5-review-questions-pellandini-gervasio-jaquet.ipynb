## a) Re-explain in your own words the steps of the K-Means algorithm.

The centroids are first initialized at random, and each data point is subsequently assigned to its closest centroid. Then, we update each centroid to the average position of the points that are allocated to it. This process is repeated again until the centroids stabilize or the distortion change becomes negligible.

## b) Are we guaranteed to observe a decreasing distortion J from one epoch to the other in the K-Means ?

Yes, the distortion in K-Means will either diminish or stay constant with each iteration. This is due to the fact that each centroid update and assignment step is intended to reduce the objective function, ensuring convergence, even though it can converge to a local minimum rather than the global one.

## c) For two different initial values of the centroids in the K-Means, can we get different end values of the distortion J ? Argument your answer.

Yes, we can get distinct end values. As demonstrated on page 42 of the course materials, K-Means can converge to several local minima based on the initial positions of the centroids.


## d) Can the K-Means be used as a compression algorithm ? Compute the compression ratio for a setting with 256 centroids and an input space at two dimensions ($x_1$, $x_2$) encoded in float32.

Yes, K-Means can indeed compress data by replacing each data point with the index of its closest centroid (reducing the overall data size).

- Each data point is represented by two 32-bit floats -> 2 x 32 = 64 bits in the orignal data.
- After compression, each point can be represented by 8-bit index.
- 64 / 8 = 8 meaning the compressed data is one-eighth the size of the original.

## e) What is the use of the elbow method ? Explain it in your own words.

This method helps determine the optimal number of clusters by finding the point where adding more clusters no longer significantly reduces distortion. We test a range of K values, plot the results, and determine the elbow point, the point at which increasing K further just slightly improves the situation.
